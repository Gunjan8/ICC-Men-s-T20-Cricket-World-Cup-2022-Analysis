```python
import pandas as pd
import json
import os
```

```python
with open('C:\\Users\\Gunjan Chandak\\Desktop\\cricket project\\t20_json_files\\t20_wc_match_results.json') as f:
    data = json.load(f)
df_match = pd.DataFrame(data[0]['matchSummary']) 
df_match.head()
```

```python
# Check no. of element (rows,column)
df_match.shape
```

```python
# To make scorecard as a match Id so later tables can be linked
df_match.rename({'scorecard':'match_id'}, axis=1, inplace=True)
df_match.head()
```

```python
# Make dictionary to connect match_id with teams so we can connect tables using that
match_ids_dict={}

for index, row in df_match.iterrows():
    key1= row['team1']+ ' Vs ' +row['team2']
    key2= row['team2']+ ' Vs ' +row['team1']
    match_ids_dict[key1] = row['match_id']
    match_ids_dict[key2] = row['match_id']

match_ids_dict
    
```

```python
df_match.to_csv('C:\\Users\\Gunjan Chandak\\Desktop\\cricket project\\dim_match_summary.csv', index = False)
    
```

```python
#batting summary
with open('C:\\Users\\Gunjan Chandak\\Desktop\\cricket project\\t20_json_files\\t20_wc_batting_summary.json') as f:
    data = json.load(f)
all_records=[]

for rec in data:
    all_records.extend(rec['battingSummary'])

df_batting = pd.DataFrame(all_records)
df_batting.head()
    
```

```python
#Convert dismissal column into out/not_out column
df_batting["out/not_out"] = df_batting.dismissal.apply(lambda x: "out" if len(x)> 0 else "not_out")
df_batting.head()
```

```python
# Now we don't need dismissal column so delete it
df_batting.drop(columns =["dismissal"],inplace=True)
df_batting.head(11)
```

```python
df_batting.head()
```

```python
# delete special characters from batsman name
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('â€',''))
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('\xa0',''))
df_batting.head()
```

```python
match_ids_dict["Namibia Vs Sri Lanka"]
```

```python
#Create new column match_id in this dataframe to connect tables
df_batting["match_id"]= df_batting["match"].map(match_ids_dict)
df_batting.head()
```

```python
df_batting.shape
```

```python
df_batting.to_csv('C:\\Users\\Gunjan Chandak\\Desktop\\cricket project\\t20_csv_files\\bating_summary.csv', index = False)
```
